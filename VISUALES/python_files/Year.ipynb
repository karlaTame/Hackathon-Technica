```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dates = pd.date_range(start='10/10/2020', periods=360, freq="D")
sales_tot=np.random.rand(360, 1)

k = pd.DataFrame(sales_tot,
                    index=dates, columns=['Ventas'])
```

```python
#Plot for a year vilualization

from pylab import rcParams

k1=np.array(k)
m=np.median(k1)
me=np.mean(k1)
k.plot()
rcParams['figure.figsize'] = 10, 10

plt.title('Sells behaviour',fontsize=24)
plt.xlabel('month',fontsize=18)
plt.ylabel('normalized sells',fontsize=18)
plt.axhline(me, color='y', linestyle='dashdot')
plt.axhline(m, color='m', linestyle='dashdot')
```
