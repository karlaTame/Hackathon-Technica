```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pylab import rcParams
```

```python
dates = pd.date_range(start='10/10/2020', periods=360, freq="D")
sales_tot=np.random.rand(360, 1)

k = pd.DataFrame(sales_tot,
                    index=dates, columns=['Sales'])
```

```python
sales_tot=np.array(sales_tot).ravel()
sales_month=np.split(sales_tot, 12);

#Monthly data division

alm=input("Enter the month ex. March")
if alm == "January":
    fq=sales_month[0]
    #print(fq)
elif alm =="February":
    fq=sales_month[1]
    #print(fq)
elif alm=="March":
    fq=sales_month[2]
    #print(fq)
elif alm =="April":
    fq=sales_month[3]
    #print(fq)
elif alm =="May":
    fq=sales_month[4]
    #print(fq)
elif alm =="June":
    fq=sales_month[5]
    #print(fq)    
elif alm =="July":
    fq=sales_month[6]
    #print(fq)
elif alm =="August":
    fq=sales_month[7]
    #print(fq)  
elif alm =="September":
    fq=sales_month[8]
    #print(fq)    
elif alm =="October":
    fq=sales_month[9]
    #print(fq)
elif alm =="November":
    fq=sales_month[10]
    #print(fq)
elif alm =="December":
    fq=sales_month[11]
    #print(fq)   
#weekly data division

wish=input("specific week?")
if wish=="yes":

    almw=input("week")
    sales_week=np.split(fq, 5);

    if almw == "week1":
        fq=sales_week[0]
        #print(fq)
    elif almw =="week2":
        fq=sales_week[1]
        #print(fq)
    elif almw=="week3":
        fq=sales_week[2]
        #print(fq)
    elif almw =="week4":
        fq=sales_week[3]
        #print(fq)
    elif almw =="week5":
        fq=sales_week[4]
        #print(fq)
    elif wish=="no":
        print("Ok")
```

```python
fq=fq.ravel()
v1=fq
pp=[]
pn=[]
qi=0
for i in v1:
    qi=qi+1
    qj=0
    for j in v1:
        qj=qj+1
        if qj==qi+1:
            if j-i < 0:
                pn.append(i)
                pn.append(j)
            elif j-i > 0:
                pp.append(i)
                pp.append(j)


up=[]
un=[]
   
for i in v1:
    for j in pp:
        if i==j:
            up.append(np.where(v1== i))
for i in v1:
    for j in pn:
        if i==j:
            un.append(np.where(v1== i)) 

un=np.array(un).ravel()+1
up=np.array(up).ravel()+1

m=np.median(v1)
me=np.mean(v1)
```

```python
x=up
a=un
y=pp
b=pn

plt.plot(a,b, 'go')

def connectpoints(a,b,p1,p2):
    a1, a2 = a[p1], a[p2]
    b1, b2 = b[p1], b[p2]
    plt.plot([a1,a2],[b1,b2],'r-')

for i in range(0,len(a),2):
    connectpoints(a,b,i,i+1)


plt.plot(x,y, 'go')

def connectpoints(x,y,p1,p2):
    x1, x2 = x[p1], x[p2]
    y1, y2 = y[p1], y[p2]
    plt.plot([x1,x2],[y1,y2],'b-')

for i in range(0,len(x),2):
    connectpoints(x,y,i,i+1)
    
if len(fq) == 30:
    plt.title('Sales behaviour',fontsize=24)
    plt.xlabel('Day of the month'+ ":" + alm,fontsize=18)
    plt.ylabel('Normalized sales',fontsize=18)  
    rcParams['figure.figsize'] = 10, 10
    plt.axhline(me, color='y', linestyle='dashdot')
    plt.axhline(m, color='m', linestyle='dashdot')
    plt.xlim(1, 30)
    plt.show()
else:
    plt.title('Sales behaviour', fontsize=24)
    plt.xlabel('Day of the week'+":"+almw,fontsize=18)
    plt.ylabel('Normalized sales',fontsize=18)  
    rcParams['figure.figsize'] = 10, 10
    plt.axhline(me, color='y', linestyle='dashdot')
    plt.axhline(m, color='m', linestyle='dashdot')
    plt.xlim(0.8, 6.2)
    plt.show()
    
```
